```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import numpy as np

# Set random seed
np.random.seed(100)

# Generate the dataset
X, y = make_moons(n_samples=250, noise=0.05, random_state=42)

# Apply KMeans
kmeans = KMeans(n_clusters=15, random_state=100)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Apply GMM
gmm = GaussianMixture(n_components=4, random_state=100)
gmm.fit(X)
y_gmm = gmm.predict(X)

# Plotting
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.title('KMeans Clustering')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_gmm, cmap='viridis')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], s=300, c='red')
plt.title('Gaussian Mixture Model')

plt.show()
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN, SpectralClustering
import matplotlib.pyplot as plt
import numpy as np

# Set random seed
np.random.seed(100)

# Generate the dataset
X, y = make_moons(n_samples=250, noise=0.05, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3)
y_dbscan = dbscan.fit_predict(X)

# Apply Spectral Clustering
spectral = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', n_neighbors=10)
y_spectral = spectral.fit_predict(X)

# Plotting
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis')
plt.title('DBSCAN Clustering')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_spectral, cmap='viridis')
plt.title('Spectral Clustering')

plt.show()
```
